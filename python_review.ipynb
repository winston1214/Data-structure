```python
a = int(input("  : "))
for i in range(a):
    print(" " * (2*(a - i)-1), end=" ")
    for j in range(i+1):
        print(2 * j + 1, end=" ")
    for j in range(i, 0, -1):
        print(2*j - 1, end=' ')
    print("\n")
```

```python
class Bag:
    def __init__(self,ls=[],string=''):
        self.ls = []
        self.string=''
    def insert(self,string):
        self.ls.append(string)
    def remove(self,string):
        self.ls.remove(string)
    def bag(self):
        return self.ls
```
