```python
r= int(input("소득(만원단위 입력) : "))
a=1200*0.06
b=3400*0.15
c=4200*0.24
d=6200*0.35
if r>15000:
    print("세금은: ",a+b+c+d+(r-15000)*0.38)
elif r>8800:
    print("세금은: ",a+b+c+(r-8800)*0.35)
elif r>4600:
    print("세금은: ",a+b+(r-4600)*0.24)
elif r>1200:
    print("세금은: ",a+(r-1200)*0.15)
else:
    print("세금은: ",r*0.06)
```

```python
import random
answer = random.randint(10,99)
guess = int(input("숫자를 입력하시오(두자리수): "))
for i in range(10):
    if answer<guess:
        print("아닙니다 더 작은 숫자입니다")
        guess = int(input("숫자를 입력하시오(두자리수): "))
    if answer>guess:
        print("아닙니다 더 큰 숫자입니다")
        guess = int(input("숫자를 입력하시오(두자리수): "))
    else:
        break
print("정답입니다. {}번 만에 숫자 {}를 맞추셨습니다.".format(i+1,answer))
```

```python
a = int(input("피라미드의 높이를 입력하세요: "))
for i in range(a):
    print(" " * (2*(a - i)-1), end=" ")
    for j in range(i+1):
        print(2 * j + 1, end=" ")
    for j in range(i, 0, -1):
        print(2*j - 1, end=' ')
    print("\n")
```

```python
class Bag:
    def __init__(self,ls=[],string=''):
        self.ls = []
        self.string=''
    def insert(self,string):
        self.ls.append(string)
    def remove(self,string):
        self.ls.remove(string)
    def bag(self):
        return self.ls
```

```python
mybag = Bag()
mybag.insert("휴대폰")
mybag.insert("지갑")
mybag.insert("손수건")
mybag.insert("빗")
mybag.insert("자료구조")
mybag.insert("야구공")
print("내 가방 속의 물건 :",mybag.bag())
mybag.insert("빗")
mybag.remove("손수건")
print("내 가방 속의 물건 :",mybag.bag())
```
